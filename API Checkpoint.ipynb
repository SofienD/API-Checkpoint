**4-Extract data from the this API using json**

```python
import json
```

```python
response = requests.get(
    'https://api.github.com/search/repositories',
    params={'q': 'requests+language:python'},
)
```

```python
json_response = response.json()
repository = json_response['items'][0] # Inspect some attributes of the `requests` repository
print(f'Repository name: {repository["name"]}')  # Python 3.6+
print(f'Repository description: {repository["description"]}')  # Python 3.6+
data= json.loads(response.text)
```

**5-Create a dataframe containing your final data from your obtained json data**

```python
import pandas as pd
df=pd.DataFrame(data)
```
